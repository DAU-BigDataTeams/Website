```python
import osmnx as ox
import networkx as nx
import pandas
```

```python
# 하단동의 중심 좌표
lat, lng = 35.139835, 129.038118

# 하단동 주변 1km 반경 내의 도로 네트워크를 가져옵니다.
G = ox.graph_from_point((lat, lng), network_type='drive')

# 교외성을 계산합니다.
bc = nx.betweenness_centrality(G)

# 교외성이 가장 높은 상위 10개 도로를 가져옵니다.
top10 = sorted(bc.items(), key=lambda x: x[1], reverse=True)

# 도로 네트워크와 교외성이 가장 높은 상위 10개 도로를 시각화합니다.
fig, ax = ox.plot_graph(G, edge_color='gray', node_size=0, show=False, close=False)
fig, ax = ox.plot_graph(G, edge_color=ox.plot.get_colors(n= len(top10),cmap='plasma'), node_size=0, edge_linewidth=2, ax=ax, show=True)
```
